```python
data_filtered = pd.read_excel(r'model_training/data/RAW/data_EDA_wiek.xlsx')
```

```python
data_filtered['AgeDays'] = data_filtered['Wiek']
```

```python
v = 'AgeDays'
```

```python
data_mean, data_std = data_filtered[v].mean(), data_filtered[v].std()
cut_off = data_std * 3
lower, upper = data_mean - cut_off, data_mean + cut_off
data_plot = data_filtered[(data_filtered[v] > lower) &
                          (data_filtered[v] < upper)]
df_trimmed = data_plot[[v, 'Label']]
```

```python
fig = sns.displot(data=df_trimmed, x=v, hue="Label", kind="kde")
fig.set(ylabel='Gęstość')
```
